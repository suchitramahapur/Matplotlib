```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpoints = np.array([0,5])
ypoints = np.array([0,20])
plt.plot(xpoints,ypoints)
plt.show()
```

```python

x = np.linspace(4,15,100)
y = np.sin(x)
plt.figure(figsize=(6,3))
plt.title("Lineplot")
plt.xlabel("Xvalues")
plt.ylabel("yvalues")
plt.grid(True)
plt.plot(x,y,"g--")
plt.show()
```

```python
x = np.array(["A","B","C","D","E"])
y = np.array([4,6,8,6,4])
plt.figure(figsize=(6,3))
plt.title("Bar Char")
plt.xlabel("xvalues")
plt.ylabel("yvalues")
plt.bar(x,y,label = "Bar",color = "purple")
plt.legend()
plt.show()
```

```python
data = np.random.rand(1000)
plt.figure(figsize = (6,4))
plt.title("Histogram")
plt.xlabel("xvalues")
plt.ylabel("yvalues")
plt.hist(data,histtype = "bar",bins=25, color = "hotpink", edgecolor ="black", alpha=0.9)
plt.show()
```

```python
size =np.array([34,56,21,69])
labels = np.array(["Ram","Sham","Sita","Gita"])
plt.figure(figsize=(6,4))
plt.pie(size,labels=labels,explode= [0.0,0.0,0.1,0.0], autopct = "%1.1f%%", colors=["r","g","y","b"])
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
size = 10000 * np.random.rand(100)

plt.figure(figsize=(6,4))
plt.title("Scatter plot")
plt.xlabel("xvalues")
plt.ylabel("ylabels")
plt.scatter(x,y,c=colors,s=size,cmap="plasma",alpha=0.6)
plt.colorbar(label = "intensity")
plt.show()
```

```python
#polar plot
angle = np.linspace(0, 2*np.pi,100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize = (6,5))
ax = plt.subplot(111, polar=True)
ax.plot(angle, radii, color="orange", linewidth=2)
ax.fill(angle, radii, color="orange", alpha=0.6)
plt.title("Polar plot")
plt.show()
```

```python
overs = np.arange(1, 21)
ind = np.random.choice(range(0,25),20)
nz = np.random.choice(range(0,25),20)
```

```python
plt.figure(figsize=(20,5))
#line plot
plt.subplot(131)
plt.plot(ind,color="blue",label="IND")
plt.plot(nz,color="black",label="NZ")
plt.legend()
# plt.show()

#bar chart
plt.subplot(132)
plt.bar(overs+1, ind, color="blue", label="IND")
plt.bar(overs+1, nz, color="black", label="NZ")
plt.legend()
# plt.show()

# Scatter
plt.subplot(133)
plt.scatter(ind,nz)
# plt.legend()
plt.show()
```
